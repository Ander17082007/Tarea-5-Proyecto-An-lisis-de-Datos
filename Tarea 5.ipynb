```python
#Datos Faltantes
df['Age'] = df['Age'].fillna(df['Age'].mean())
valor_mas_frecuente = df['Cabin'].mode()[0]
df['Cabin'] = df['Cabin'].fillna(valor_mas_frecuente)
valor_mas_frecuente = df['Embarked'].mode()[0]
df['Embarked'] = df['Embarked'].fillna(valor_mas_frecuente)
```

```python
#Verificación de los datos faltantes
print("Número de registros con edad igual a 0: ", df["Age"].isin([0]).sum())
print("Número de registros conn 0 en la embarcación ", df["Embarked"].isin([0]).sum())
print("Número de registros con valor 0 en la tarifa: ", df["Fare"].isin([0]).sum())
print("Número de registros con 0 en la clase PC", df["Pclass"].isin([0]).sum())
print("Número de registros con 0 en la cabina ", df["Cabin"].isin([0]).sum())
```

```python
#Reemplazamos los valores de 0 con el promedio de cada columna 
df['Fare']=df['Fare'].replace(0,df['Fare'].mean())
```

```python
#Definimos los niveles para poner en consideración los valores atipicos
nivel_minimo=1
nivel_maximo=300
Total_Atipicos=((df['SibSp'] < nivel_minimo) | (df['SibSp'] > nivel_maximo)).sum()
print("La cantidad total de datos atipicos es: {}".format(Total_Atipicos))
```

```python
#Filtamos los datos atipicos presentes
df = df[(df['SibSp'] <= nivel_maximo) & (df['SibSp'] >= nivel_minimo)]
```

```python
#Validamos los datos
Total_Atipicos=((df['SibSp'] < nivel_minimo) | (df['SibSp'] > nivel_maximo)).sum()
print("La cantidad total de datos atipicos es: {}".format(Total_Atipicos))
```

```python
#Total de personas que lograron sobrevivir o no al naufragio
df['Survived'].value_counts()
```

```python
#División en Train y Test
# Porcentaje de datos asignados para el conjunto de entrenamiento (train) = 75%, y para el conjunto de prueba (test) = 25%
# Se definen las variables X e Y para guardar la información del conjunto de entrenamiento
# Se cargan los datos de entrenamiento excluyendo la columna de sobrevivientes
X = df.drop(["Survived"], axis=1)  
# Se cargan los datos de la columna de sobrevivientes
Y = df["Survived"]  
# La función train_test_split realiza una partición de los datos en dos grupos: uno para entrenamiento y otro para prueba (train y test)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, stratify=Y, random_state=0)
```

```python
print(X_train)
```

```python
print(Y_train)
```

```python
# Crear el modelo de regresión logística
modelo_reglog = LogisticRegression(solver="liblinear")
```

```python
# Ajustar el modelo con los datos de entrenamiento
modelo_reglog.fit(X_train, Y_train)  
```

```python
# Crear el modelo de predicción
prediccion = modelo_reglog.predict(X_test)
```

```python
# Comprobamos los resultados reales del conjunto de prueba
print(Y_test)
```

```python
print(prediccion)
```

```python
#Evaluamos el desempeño del modelo
print(classification_report(Y_test,prediccion))
```

```python
#Gráficas que evidencian el correspondiente resultado (Matriz de Confusión)
cm=confusion_matrix(Y_test,prediccion)
disp=metrics.ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
#Gráficas que evidencian el correspondiente resultado (Curva de Precisión)
from sklearn.metrics import RocCurveDisplay
RocCurveDisplay.from_predictions(Y_test,prediccion)
plt.show()
```

```python
#Gráficas que evidencian el correspondiente resultado (Matriz de confusión normalizada)
from sklearn.metrics import ConfusionMatrixDisplay
cm_normalized = confusion_matrix(Y_test, prediccion, normalize='true')
disp_normalized = ConfusionMatrixDisplay(confusion_matrix=cm_normalized)
disp_normalized.plot(cmap='Blues')
plt.title('Matriz de Confusión Normalizada')
plt.show()
```

```python
#Gráficas que evidencian el correspondiente resultado (Curva de precisión-recall)
from sklearn.metrics import PrecisionRecallDisplay
disp_pr = PrecisionRecallDisplay.from_predictions(Y_test, prediccion)
disp_pr.plot()
plt.title('Curva de Precisión-Recall')
plt.show()
```
